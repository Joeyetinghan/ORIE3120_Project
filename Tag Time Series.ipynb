```python
def remove_url(txt):
    """Replace URLs found in a text string with nothing 
    (i.e. it will remove the URL from the string).

    Parameters
    ----------
    txt : string
        A text string that you want to parse and remove urls.

    Returns
    -------
    The same txt string with url's removed.
    """

    return " ".join(re.sub("([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "", txt).split())
```

```python
df['tag_list'] = df['tags'].str.split(pat="|")
```

```python
df_music = df[df['categoryId'] == 10]
def count(tag_list, target_tag):
    count = 0
    for tag in tag_list:
        tags = remove_url(tag).lower().split()
        for x in tags:
            if target_tag == x:
                count +=1
    return count
df_music['tag_count_rap'] = df_music['tag_list'].apply(lambda x: count(x, 'rap'))

df_music
```
